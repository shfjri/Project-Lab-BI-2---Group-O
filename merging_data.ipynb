```python
import pandas as pd
import os
from tqdm import tqdm
```

```python
sales_files = [f'./dataset/Sales/{file}' for file in os.listdir('./dataset/Sales')]
sales_files
```

```python
cust_df = pd.read_csv(sales_files[0]).drop_duplicates().reset_index(drop=True)
cust_df.head()
```

```python
orders_df = pd.read_csv(sales_files[1]).drop_duplicates().reset_index(drop=True)
orders_df.head()
```

```python
items_df = pd.read_csv(sales_files[2]).drop_duplicates().reset_index(drop=True)
items_df.head()
```

```python
review_df = pd.read_csv(sales_files[3]).drop_duplicates().reset_index(drop=True)
review_df.head()
```

```python
product_df = pd.read_csv(sales_files[4]).drop_duplicates().reset_index(drop=True)
product_df.head()
```

```python
seller_df = pd.read_csv(sales_files[5]).drop_duplicates().reset_index(drop=True)
seller_df.head()
```

Join Order data and Items data

```python
df1 = orders_df.merge(items_df)
df1
```

Join again with Product data

```python
df1 = df1.merge(product_df)
df1
```

Join merged dataframe above and Review data

```python
df1 = df1.merge(cust_df)
df1
```

Join merged dataframe above and Order Item data

```python
df1 = df1.merge(seller_df)
df1
```

Join merged dataframe above and Product data

```python
df1 = df1.merge(review_df)
df1.head()
```

Save merged dataframe as raw data final

```python
df1.to_csv('./dataset/ready/raw_data_final.csv',index=False)
```
